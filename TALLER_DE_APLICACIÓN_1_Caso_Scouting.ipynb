# 3. Un miembro del cuerpo técnico asegura que los arqueros (GK) altos tienen un mejor  desempeño (overall). ¿Qué puede afirmar usted a través de un análisis descriptivo?  (Nota: tome por definición tres grupos de estatura: i) menor a 185 cm; ii) entre 185 y  192 cm; iii) más de 192 cm)

```python
# Filtrar solo los arqueros (GK)
gk_data = fifa[fifa["position"] == "GK"]

# Crear los grupos de altura
bins = [0, 185, 192, float("inf")]
labels = ["Bajo (<185 cm)", "Mediano (185-192 cm)", "Alto (>192 cm)"]
gk_data.loc[:,"height_group"] = pd.cut(gk_data["height_cm"], bins=bins, labels=labels)

# Calcular estadísticas descriptivas por grupo
stats = gk_data.groupby("height_group")["overall"].agg(["count", "mean", "median", "std"])
print(stats)

# Agrupar los valores de "overall" por cada grupo de altura
grouped_overall = [gk_data[gk_data["height_group"] == label]["overall"] for label in labels]

# Visualizar con un boxplot usando Matplotlib
plt.figure(figsize=(8, 5))
plt.boxplot(grouped_overall, labels=labels)

plt.title("Distribución del Overall por Grupo de Altura (GK)")
plt.xlabel("Grupo de Altura")
plt.ylabel("Overall")

# Mostrar el gráfico
plt.show()
```

## Tendencia Central  

- Los arqueros más bajos (<185 cm) tienen una **media** de **62.09**, mientras que los medianos (185-192 cm) y altos (>192 cm) tienen medias de **64.48** y **64.75**, respectivamente.  
- La **mediana** sigue una tendencia similar, con **62** para los bajos y **64** para los medianos y altos.  

## Dispersión de los Datos  

- La **desviación estándar** es similar entre los grupos (**≈7.3 a 7.6**), lo que indica que la variabilidad del overall es parecida en todos los casos.  
- Se observan **valores atípicos (outliers)** en todos los grupos, pero no parecen afectar demasiado la distribución general.  

## Boxplot (Gráfico de Cajas)  

- El grupo de **arqueros bajos** tiene una distribución más centrada en valores más pequeños de overall.  
- Los grupos **medianos y altos** tienen distribuciones muy similares, con **medianas idénticas** y **rangos parecidos**.  

## Conclusiones  

- Los arqueros más bajos (<185 cm) tienen un **overall ligeramente menor** en promedio que los medianos y altos.  
- Los arqueros **medianos y altos tienen prácticamente el mismo overall**.  
- Si el objetivo es elegir un arquero con mejor desempeño (**overall**), los datos sugieren que **ser más alto de 185 cm es una ventaja**, pero no hay una gran diferencia entre los arqueros medianos y altos.

 # 5. Un scout asegura que los mejores jugadores, según el puntaje de desempeño (overall), juegan en clubes que pertenecen al grupo de “ligas top5” (Inglaterra, España,  Alemania, Italia y Francia). ¿Qué puede decir sobre estas dos afirmaciones a través de  un análisis descriptivo?  


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Definir las ligas top 5
top5_countries = ["England", "Spain", "Germany", "Italy", "France"]

# Crear una nueva columna que indique si el jugador pertenece a una liga top 5
fifa["is_top5"] = fifa["club_country"].isin(top5_countries)

# Calcular estadísticos descriptivos
top5_overall = fifa[fifa["is_top5"]]["overall"]
other_overall = fifa[~fifa["is_top5"]]["overall"]

# Media, mediana y moda
print("\nEstadísticas de jugadores en ligas Top5:")
print(f"Media: {top5_overall.mean():.2f}")
print(f"Mediana: {top5_overall.median():.2f}")
print(f"Moda: {top5_overall.mode().values[0]}")
print(fifa[fifa["is_top5"]]["overall"].describe())

print("\nEstadísticas de jugadores en otras ligas:")
print(f"Media: {other_overall.mean():.2f}")
print(f"Mediana: {other_overall.median():.2f}")
print(f"Moda: {other_overall.mode().values[0]}")
print(fifa[~fifa["is_top5"]]["overall"].describe())

# Visualización: Boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(x=fifa["is_top5"], y=fifa["overall"])
plt.xticks([0, 1], ["Otras ligas", "Top5"])
plt.xlabel("Grupo de Ligas")
plt.ylabel("Puntaje Overall")
plt.title("Distribución del Overall en Ligas Top5 vs. Otras Ligas")
plt.show()

# Visualización: Histogramas
plt.figure(figsize=(10, 5))
sns.histplot(top5_overall, kde=True, color="blue", label="Top5")
sns.histplot(other_overall, kde=True, color="red", label="Otras ligas")
plt.legend()
plt.xlabel("Puntaje Overall")
plt.ylabel("Frecuencia")
plt.title("Distribución de Overall en Ligas Top5 vs. Otras Ligas")
plt.show()
```

# 6. Otro scout asegura que la proporción de jugadores zurdos es mayor en la posición de  centro campista (CM) que en la de centro campista ofensivo (CAM). ¿Qué puede decir  sobre esta afirmación?

```python
# Filtrar jugadores por posición
cm_players = fifa[fifa["position"] == "CM"]
cam_players = fifa[fifa["position"] == "CAM"]

# Calcular cantidad total y zurdos en cada posición
cm_total = len(cm_players)
cam_total = len(cam_players)

cm_left = len(cm_players[cm_players["pref_foot"] == "Left"])
cam_left = len(cam_players[cam_players["pref_foot"] == "Left"])

# Calcular proporciones
cm_proportion = cm_left / cm_total
cam_proportion = cam_left / cam_total

# Mostrar resultados
print(f"Proporción de zurdos en CM: {cm_proportion:.2%}")
print(f"Proporción de zurdos en CAM: {cam_proportion:.2%}")

# Comparación
if cm_proportion > cam_proportion:
    print("✅ Hay una mayor proporción de jugadores zurdos en la posición CM que en CAM.")
else:
    print("❌ No hay una mayor proporción de jugadores zurdos en CM que en CAM.")
```

```python
cm_players = fifa[fifa['position'] == 'CM']
cam_players = fifa[fifa['position'] == 'CAM']

# Gráfico de pastel para CM
plt.figure(figsize=(8, 8))
cm_colors = ['royalblue', 'lightgray']
cm_players['pref_foot'].value_counts().plot.pie(autopct='%1.1f%%', colors=cm_colors, startangle=140, wedgeprops={'edgecolor': 'black'})
plt.title('Distribución de pie preferido en CM', fontsize=14, fontweight='bold')
plt.ylabel('')
plt.show()

# Gráfico de pastel para CAM
plt.figure(figsize=(8, 8))
cam_colors = ['royalblue', 'lightgray']
cam_players['pref_foot'].value_counts().plot.pie(autopct='%1.1f%%', colors=cam_colors, startangle=140, wedgeprops={'edgecolor': 'black'})
plt.title('Distribución de pie preferido en CAM', fontsize=14, fontweight='bold')
plt.ylabel('')
plt.show()
```

# 7. Otro scout asegura que los mejores jugadores juegan en los mejores clubes del mundo  ¿Es cierta esta afirmación de que existe una correlación entre el desempeño de los  jugadores (overall) y el puntaje de desempeño del club al que pertenece  (club_overall)?


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Calcular la correlación de Pearson
corr, p_value = pearsonr(fifa['overall'], fifa['club_overall'])
print(f"📊 Correlación de Pearson: {corr:.4f} (p-valor: {p_value:.4e})")

# Visualizar la relación con un gráfico de dispersión y línea de regresión
plt.figure(figsize=(8, 5))
sns.regplot(x=fifa['overall'], y=fifa['club_overall'], scatter_kws={'alpha': 0.5}, line_kws={'color': 'red'})
plt.title(f'Relación entre Overall del Jugador y Club Overall\nCorrelación: {corr:.2f}')
plt.xlabel('Puntaje Overall del Jugador')
plt.ylabel('Puntaje Overall del Club')
plt.show()
```

## **Punto 7**
# Análisis de Correlación entre Overall del Jugador y Club Overall  

A través de la **correlación de Pearson**, el gráfico de **dispersión con regresión lineal** muestra la relación entre el **overall** de los jugadores y el **club_overall**.  

## Resultados de la Correlación  

- La correlación de **0.58** sugiere una **relación moderadamente fuerte** entre ambas variables.  
- Esto significa que los jugadores con **mejor desempeño** suelen estar en clubes con un **mayor puntaje**, aunque **no siempre**.  
- El **p-valor cercano a 0.0000** indica que esta correlación es **significativa** y no producto del azar.  

## Excepciones Observadas  

Existen casos que no siguen esta tendencia general:  
1. **Jugadores con alto overall en clubes de menor nivel**.  
2. **Jugadores con menor overall en equipos élite**.  

## Interpretación del Diagrama de Dispersión  

- La **línea de regresión** muestra la **tendencia general**.  
- Sin embargo, la **dispersión de los puntos** sugiere que otros factores, como **fichajes o la edad del jugador**, también influyen en la distribución.  

## Conclusión  

El scout tiene razón **en general**, pero **no es una regla absoluta**.
